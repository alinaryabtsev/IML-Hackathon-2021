```python

last_year_df = prep.get_df()#[prep.get_df()['year'==2021]]
a = last_year_df['year'] == 2020
b = last_year_df['year'] == 2021
c = a + b

last_year_df = last_year_df[c]
# print(last_year_df[['release_date']].dtypes)
# last_year_df['today'] = pd.Timestamp('today').strftime("%Y-%m-%d")
last_year_df['today'] = pd.to_datetime(pd.to_datetime('today').strftime("%Y-%m-%d"))
print(last_year_df['today'].dtype)
last_year_df['days_from_2021']= (last_year_df['release_date'] - last_year_df['today'] ).dt.days
print(last_year_df.shape)
last_year_df=last_year_df[last_year_df['days_from_2021']<0]
print(last_year_df.shape)
last_year_df[['revenue','vote_average','days_from_2021']].head(40)
```

```python
print(last_year_df['days_from_2021'].dtype)
```

```python
# Plot the data
f, ax = plt.subplots()

ax.bar(x=last_year_df.days_from_2021,
       height=last_year_df.revenue,
       color="purple")

ax.set(title="Plot of revenue by last days")
plt.show()
```

```python
# Plot the data
f, ax = plt.subplots()

ax.bar(x=last_year_df.days_from_2021,
       height=last_year_df.vote_average,
       color="purple")

ax.set(title="Plot of vote_average by last days")
plt.show()
```
